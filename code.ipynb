```python
from pypdf import PdfReader, PdfWriter

# Settings: Extracts pages 4 through 8 (inclusive)
input_file, output_file, start_page, end_page = "FYJC-Chemistry.pdf", "test2.pdf", 12, 13

reader = PdfReader(input_file)
writer = PdfWriter()

# Loop through the specific range (subtracting 1 because Python starts at index 0)
for i in range(start_page - 1, end_page):
    writer.add_page(reader.pages[i])

writer.write(output_file)
print("PDF created successfully!")
```
